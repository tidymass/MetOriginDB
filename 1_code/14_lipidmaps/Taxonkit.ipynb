```python
import pandas as pd
import re
```

```python
lipid_data = pd.read_csv("/Users/ejo/Desktop/lipid.csv")
lipid_data
```

```python
lipid_data['ncbi_id'] = lipid_data['Curated.from'].apply(
    lambda x: re.search(r'#(\d+)', str(x)).group(1) if pd.notna(x) and re.search(r'#(\d+)', str(x)) else None
)
```

```python
lipid_data
```

```python
lipid_data = lipid_data.dropna(subset=['ncbi_id'])
```

```python
lipid_data = lipid_data.head(10)
```

```python
df = lipid_data
```

```python
df
```

```python
from ete3 import NCBITaxa
```

```python
# 初始化 NCBI Taxonomy
ncbi = NCBITaxa()
```

```python
# 查询分类信息
def get_taxonomy_ete3(ncbi_id):
    try:
        lineage = ncbi.get_lineage(int(ncbi_id))  # 获取分类路径
        ranks = ncbi.get_rank(lineage)  # 获取分类级别
        names = ncbi.get_taxid_translator(lineage)  # 获取名称
        
        # 创建分类字典
        tax_dict = {ranks[taxid]: names[taxid] for taxid in lineage if ranks[taxid] in ['superkingdom', 'phylum', 'class', 'order', 'family', 'genus', 'species']}
        
        return pd.Series([tax_dict.get('superkingdom'), tax_dict.get('phylum'), tax_dict.get('class'),
                          tax_dict.get('order'), tax_dict.get('family'), tax_dict.get('genus'), tax_dict.get('species')])
    except Exception as e:
        return pd.Series([None] * 7)

# 获取分类
df[['superkingdom', 'phylum', 'class', 'order', 'family', 'genus', 'species']] = df['ncbi_id'].apply(get_taxonomy_ete3)
```
